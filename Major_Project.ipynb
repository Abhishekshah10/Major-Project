```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt
from datetime import timedelta
```

```
import pandas as pd
data = pd.read_csv('/content/COVID-19 Cases(13-12-2020).csv')
data
```

```
data.info()
```

```
# Dropping a Column
data.drop(["S. No."],1,inplace=True)
```

```
data["Date"] = pd.to_datetime(data["Date"])
```

```
data["Date"]
```

```
# Grouping Diff types of cases as per the date
datewise = data.groupby(["Date"]).agg({"Confirmed Cases":"sum","Active Cases":"sum","Cured/Discharged":"sum","Death":"sum"})
```

```
print("Total number of Confirmed cases in India",datewise["Confirmed Cases"].iloc[-1])
print("Total number of Recovered cases in India",datewise["Cured/Discharged"].iloc[-1])
print("Total number of Death cases in India",datewise["Death"].iloc[-1])
print("Total number of Active cases in India",datewise["Active Cases"].iloc[-1])
print("Total number of Closed cases in India",(datewise["Cured/Discharged"].iloc[-1]+datewise["Death"].iloc[-1]))
```

```
plt.figure(figsize=(15,5))
sns.barplot(x=datewise.index.date,y=datewise["Active Cases"])
plt.title("Distributions plot for Active Cases")
plt.xticks(rotation=90)
```

```
plt.figure(figsize=(15,5))
sns.barplot(x=datewise.index.date,y=datewise["Cured/Discharged"]+datewise["Death"])
plt.title("Distribution plot for Closed Cases")
plt.xticks(rotation=90)
```

```
datewise["WeekofYear"] = datewise.index.weekofyear
week_num = []
weekwise_confirmed = []
weekwise_cured = []
weekwise_deaths = []
weekwise_active = []
w = 1
for i in list(datewise["WeekofYear"].unique()):
    weekwise_confirmed.append(datewise[datewise["WeekofYear"]==i]["Confirmed Cases"].iloc[-1])
    weekwise_cured.append(datewise[datewise["WeekofYear"]==i]["Cured/Discharged"].iloc[-1])
    weekwise_deaths.append(datewise[datewise["WeekofYear"]==i]["Death"].iloc[-1])
    weekwise_active.append(datewise[datewise["WeekofYear"]==i]["Active Cases"].iloc[-1])
    week_num.append(w)
    w=w+1
plt.figure(figsize=(8,5))
plt.plot(week_num,weekwise_confirmed,linewidth=3)
plt.plot(week_num,weekwise_cured,linewidth =3)
plt.plot(week_num,weekwise_deaths,linewidth = 3)
plt.plot(week_num,weekwise_active,linewidth = 3)
plt.xlabel("WeekNumber")
plt.ylabel("Number of cases")
plt.title("Weekly Progress of different types of cases")
```

```
print("Average increase in number of Confirmed cases everyday:",np.round(datewise["Confirmed Cases"].diff().fillna(0).mean()))
print("Average increase in number of Recovered cases everyday:",np.round(datewise["Cured/Discharged"].diff().fillna(0).mean()))
print("Average increase in number of Death cases everyday:",np.round(datewise["Death"].diff().fillna(0).mean()))
print("Average increase in number of Active cases everyday:",np.round(datewise["Active Cases"].diff().fillna(0).mean()))

plt.figure(figsize=(15,6))
plt.plot(datewise["Confirmed Cases"].diff().fillna(0),label="Daily increase in confirmed cases",linewidth=3)
plt.plot(datewise["Cured/Discharged"].diff().fillna(0),label="Daily increase in recovered cases",linewidth=3)
plt.plot(datewise["Death"].diff().fillna(0),label="Daily increase in death cases",linewidth=3)
plt.plot(datewise["Active Cases"].diff().fillna(0),label="Daily increase in active cases",linewidth=3)
plt.xlabel("Timestamp")
plt.ylabel("Daily increase")
plt.title("Daily increase")
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```
#State wise analysis
#Calculating State wise Mortality rate
statewise= data[data["Date"]==data["Date"].max()].groupby(["Region"]).agg({"Confirmed Cases":"sum","Cured/Discharged":"sum","Death":"sum","Active Cases":"sum"}).sort_values(["Confirmed Cases"],ascending=False)
statewise["Mortality"]=(statewise["Death"]/statewise["Cured/Discharged"])*100
statewise["Cured/Discharged"]=(statewise["Cured/Discharged"]/statewise["Confirmed Cases"])*100
```

```
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(25,10))
top_15confirmed = statewise.sort_values(["Confirmed Cases"],ascending=False).head(15)
top_15deaths = statewise.sort_values(["Death"],ascending=False).head(15)
sns.barplot(x=top_15confirmed["Confirmed Cases"],y=top_15confirmed.index,ax=ax1)
ax1.set_title("Top 15 states as per number of confirmed cases")
sns.barplot(x=top_15deaths["Death"],y=top_15deaths.index,ax=ax2)
ax2.set_title("Top 15 states as per number of death cases")
```
